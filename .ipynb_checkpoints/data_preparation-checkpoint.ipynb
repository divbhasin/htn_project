```python
import pandas as pd
import json
import re
import numpy as np
%matplotlib inline
```

```python
DATA_DIR = '/Users/divbhasin/Downloads/HTN/'
```

```python
products = pd.read_csv(DATA_DIR+'products.txt', sep='\t', header=None)
products.columns = ['item', 'code', 'item_name']
products.head()
```

```python
categories = pd.read_csv(DATA_DIR+'mch_categories.tsv', sep='\t')
categories.head()
```

```python
%%time
with open(DATA_DIR+'transactions.txt', 'r') as data_file:    
    transactions = data_file.readlines()
```

```python
%%time
transactions = [json.loads(trans) for trans in transactions]
```

```python
%%time
df_transactions = pd.DataFrame([(i, trans['customer'], trans['date'], trans['store']) for i,trans in enumerate(transactions)],
                              columns=['trans_id', 'customer', 'date', 'store'])
```

```python
df_transactions.tail()
```

```python
%%time
trans_item = []
for i, trans in enumerate(transactions):
    for item in trans['itemList']:
        trans_item.append([i, item['item'], item['price'], item['quantity']])
```

```python
trans_item = pd.DataFrame(trans_item, columns=['trans_id', 'item', 'price', 'quantity'])
trans_item.tail()
```

```python
df_transactions.store.value_counts()[:10].plot(kind='barh')
```

```python
products.loc[products.item.isin(["20189092_EA", "20175355001_KG", "20668578_EA"]),]
```

```python
%%time
date_transaction=pd.merge(trans_item, df_transactions, how='left', left_on='trans_id', right_on='trans_id')
```

```python
date_transaction.tail()
```

```python
date_transaction.drop(['quantity', 'customer', 'store'], 1, inplace=True)
```

```python
from datetime import datetime
```

```python
%%time
date_transaction['date']=date_transaction['date'].apply(lambda x: datetime.strptime(x, '%Y.%m.%d %H:%M:%S'))
```
